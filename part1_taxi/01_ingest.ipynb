## Data Ingestion

#### Imports

```python
from pathlib import Path
import pandas as pd
import requests
```

#### Defining dataset URLs for downloading

```python
PARQUET_URL = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet"
LOOKUP_URL = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
```

#### Downloading and saving datasets to correct folder

```python
parquet_path = Path("../data/raw/tlc/yellow_tripdata_2023-01.parquet")
lookup_path = Path("../data/raw/lookup/taxi_zone_lookup.csv")
```

```python
def download_file(url, path):
    if path.exists():
        print(f"{path} already exists")
        return
    
    response = requests.get(url)
    response.raise_for_status()

    with open(path, "wb") as f:
        f.write(response.content)

    print(f"Downloaded {path}")
```

```python
download_file(PARQUET_URL, parquet_path)
download_file(LOOKUP_URL, lookup_path)
```

#### Loading datasets

```python
tripdata = pd.read_parquet(parquet_path)
zone_lookup = pd.read_csv(lookup_path)
```
